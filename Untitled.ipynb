```python
from tweets.model import TweetsModel
from representations.model import WordRepresentationsModel, SentenceRepresentationsModel
from representations.controller import WordRepresentationsController
from representations.generator import RepresentationsGenerator
```

```python
tweets_model = TweetsModel()
```

```python
sentence_representations_model = SentenceRepresentationsModel('sent2vec', 50)
representation_controller = WordRepresentationsController(tweets_model, sentence_representations_model)
```

```python
X = representation_controller.get_representations()
y = representation_controller.get_labels()
```

```python
X.shape, y.shape
```

```python
from sklearn.manifold import TSNE
X_embedded = TSNE(n_components=2).fit_transform(X)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
f, ax = plt.subplots(figsize=(14, 12))
sns.scatterplot(x=X_embedded[:,0], y=X_embedded[:,1], alpha=0.5)
plt.show()
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)

dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

param = {
    'max_depth': 6,
    'eta': 1,
#     'gamma': 6,
    'objective': 'binary:logistic',
    'nthread': 8,
    'eval_metric': 'auc',
    'subsample': 0.5
}
evallist = [(dtest, 'eval'), (dtrain, 'train')]

num_round = 100
bst = xgb.train(param, dtrain, num_round, evallist)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
accuracy_score(y_train, bst.predict(dtrain).round()), accuracy_score(y_test, bst.predict(dtest).round())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold

kf = KFold(n_splits=10)
for train_index, test_index in kf.split(X):
    X_train, y_train, X_test, y_test = X[train_index], y[train_index], X[test_index], y[test_index]
    clf = LogisticRegression(random_state=0, solver='lbfgs', max_iter=500)
    clf.fit(X_train, y_train)
    print('train',clf.score(X_train, y_train),'test',clf.score(X_test, y_test))
```
